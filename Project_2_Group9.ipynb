## Replacing 0 and 1 in the "Satisfaction" Column

To replace the binary values (0 and 1) in the "Satisfaction" column, we will:

1. **Calculate the Average Ratings** for the following categories:
   - `inflight_wifi_service`  
   - `departure/arrival_time_convenient`
   - `ease_of_online_booking`
   - `food_and_drink`
   - `inflight_entertainment`
   - `baggage_handling`
   - `cleanliness`

2. **Categorize Satisfaction**:
   - If the average rating across these categories is **above 3**, the passenger will be categorized as **satisfied**.
   - If the average rating is **3 or below**, the passenger will be categorized as **dissatisfied**.
